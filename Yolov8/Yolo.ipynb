```python
# Ruta del modelo y del video (ajusta)
MODEL_PATH = "rutav4/weights/best.pt"
VIDEO_PATH = "video1.mp4"
OUTPUT_PATH = "video1_yolo.mp4"

model = YOLO(MODEL_PATH)
device = "cuda" if torch.cuda.is_available() else "cpu"
model.to(device)

# ----------------------------
# 4) ParÃ¡metros de inferencia
# ----------------------------
CONF_TH = 0.7          # umbral de confianza
IMG_SZ = 800           # tamaÃ±o a pasar al modelo (opcional)
SAVE_OUTPUT = True
FPS = 20.0             # FPS para el video de salida

# ----------------------------
# 5) Abrir video y preparar writer
# ----------------------------
cap = cv2.VideoCapture(VIDEO_PATH)
width  = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))

if SAVE_OUTPUT:
    fourcc = cv2.VideoWriter_fourcc(*"mp4v")
    out = cv2.VideoWriter(OUTPUT_PATH, fourcc, FPS, (width, height))

CLASS_NAMES = ["background", "adecuado", "defectuoso"]
COLOR_OK = (0,255,0)
COLOR_BAD = (0,0,255)

frame_idx = 0
print("Iniciando inferencia en video...")

# ----------------------------
# 6) Loop inferencia por frame
# ----------------------------
while True:
    ret, frame = cap.read()
    if not ret:
        break
    frame_idx += 1
    print(f"Procesando frame {frame_idx}", end="\r")
    results = model.predict(source=frame, conf=CONF_TH, imgsz=IMG_SZ, device=device, verbose=False)

    # results es una lista (por batch); tomamos el primero
    if len(results) > 0:
        res = results[0]
        # res.boxes: contenedor con 'xyxy', 'conf', 'cls'
        boxes = res.boxes.xyxy.cpu().numpy()     # Nx4 (x1,y1,x2,y2)
        scores = res.boxes.conf.cpu().numpy()    # N
        classes = res.boxes.cls.cpu().numpy().astype(int)  # N (Ã­ndice de clase)

        # Dibujar cada caja vÃ¡lida
        for (box, score, cls_idx) in zip(boxes, scores, classes):
            # filtrado redundante por score (ya lo hizo model.predict)
            if score < CONF_TH:
                continue
            x1, y1, x2, y2 = map(int, box.tolist())
            label = CLASS_NAMES[cls_idx] if cls_idx < len(CLASS_NAMES) else f"class{cls_idx}"
            color = COLOR_OK if label == "adecuado" else COLOR_BAD

            # dibujar rectÃ¡ngulo y texto
            cv2.rectangle(frame, (x1, y1), (x2, y2), color, 2)
            txt = f"{label} {score:.2f}"
            cv2.putText(frame, txt, (x1, max(15,y1-5)), cv2.FONT_HERSHEY_SIMPLEX, 0.55, color, 2, cv2.LINE_AA)

    # Mostrar frame (Colab)
    cv2_imshow(frame)

    # Escribir en video de salida
    if SAVE_OUTPUT:
        out.write(frame)

# ----------------------------
# 7) Liberar recursos
# ----------------------------
cap.release()
if SAVE_OUTPUT:
    out.release()
print(f"\nProceso finalizado. Video guardado en: {OUTPUT_PATH}")
```
